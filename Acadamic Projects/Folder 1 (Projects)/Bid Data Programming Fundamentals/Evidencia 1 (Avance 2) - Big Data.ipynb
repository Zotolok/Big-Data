### Ejercicio 1: Número de casos confirmados por cada mes.

```python
import datetime # Manipulación de fechas
from dateutil.relativedelta import relativedelta

first_month = datetime.datetime(2020,2,1)
last_month = datetime.datetime(2023,1,1)
months = {}

while first_month <= last_month:
    m = first_month.strftime("%m-%Y")
    months[m] = 0
    first_month = first_month + relativedelta(months = 1)

print(months)
```

```python
columns = list(df.columns)
columns.remove("cve_ent")
columns.remove("poblacion")
columns.remove("nombre")
columns.remove("2020")

print(columns)
```

```python
for column in columns:
    month = column[3:] # Nomenclatura ":": Obten el valor a partir de... Ejemplo: [3:] = Obten los valores a partir del 3, tomando en cuenta el 3  
    months[month] = months[month] + df[column]
```

```python
print(months)
```

```python
pd.DataFrame(months)
```

```python
def casosConfirmadosMes():
    casos_por_mes_df = pd.DataFrame(months)
    casos_por_mes_df["nombre"] = df["nombre"]
    
    return casos_por_mes_df

casosConfirmadosMes()
```
