```python
def format_name(text):
    text = text.replace(",","")
    mylist = text.split()
    resultado = ""
    for text in mylist:
        if resultado != "":
            resultado = resultado + " "
        resultado = resultado + text.capitalize()
    return resultado
```

```python
format_name("LUIS ARTURO LEZAMA LECHAM")
```

```python
df["Nombre"] = df["Nombre"].apply(format_name) # Le ponemos igual para que el cambio se mantenga posteriormente y no solo en esta celda
df
```

```python
def calcular_calif(fila):
    total = fila["Preguntas_correctas_1"] + fila["Preguntas_incorrectas_1"]
    return fila["Preguntas_correctas_1"] / total *100

df.apply(calcular_calif, axis = 1)  # axis = 1: Evaluar por fila  axis = 0: Evaluar por columna
```
