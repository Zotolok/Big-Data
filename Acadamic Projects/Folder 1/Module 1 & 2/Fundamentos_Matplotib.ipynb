```python
import matplotlib.pyplot as plt
```

```python
import random
y = [random.randint(100,200) for x in range (len(x))]
#y2 = [random.randint(100,200) for x in range (len(x))]
```

## Line Plot

```python
plt.plot(x,y, marker="^", linestyle="--", color = "k")
plt.show()
```

```python
plt.plot(x,y, marker="^", linestyle="--", color = "b")
plt.title("Población de x país durante 2016 - 2024")
plt.show()
```

```python
plt.plot(x,y, marker="^", linestyle="--", color = "r")
plt.title("Población de x país durante 2016 - 2024")
plt.xlabel("Año")
plt.ylabel("Población (M)")
plt.show()
```

```python
plt.figure(figsize=(10,6)) # Ancho, Alto

plt.plot(x,y, marker="^", linestyle="--", color = "r")
plt.title("Población de x país durante 2016 - 2024")
plt.xlabel("Año")
plt.ylabel("Población (M)")
plt.yticks([100,150,200])

plt.savefig("Población.png")
plt.show()
```

```python
import random
y = [random.randint(100,200) for x in range (len(x))]
y2 = [random.randint(100,200) for x in range (len(x))]
```

```python
plt.figure(figsize=(10,6)) # Ancho, Alto

plt.plot(x,y, marker="^", linestyle="--", color = "r")
plt.plot(x,y2, marker="o", linestyle="-", color = "b")
plt.title("Población de x país durante 2016 - 2024")
plt.xlabel("Año")
plt.ylabel("Población (M)")
plt.yticks([100,150,200])

plt.savefig("Población.png")
plt.show()
```

## Bar Plot

```python
x = ["México", "España", "Brasil"]
y = [100, 88, 93]

plt.bar(x,y, color="r")
plt.show()
```

## Histogram

```python
import random
x = [random.randint(50,100) for x in range(32)]
bins = [50,60,70,80,90,100]

plt.hist(x,bins,edgecolor="k")
plt.show()
```

## Scatter plot - Gráfica de dispersión

## Subplots

```python
import random
x = [x for x in range(1950,2024)]
y = [ random.randint(100,200) for x in range(len(x))]
y2 = [ random.randint(100,200) for x in range(len(x))]

fig, ax = plt.subplots(1,2)
ax[0].plot(x,y, color="g")
ax[1].plot(x,y2,color="r")
plt.show()
```
