```python
# Incluya aquí su código para importar librerías y datos

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
datos = pd.read_csv("/content/drive/MyDrive/TecMilenio/DatosMasivos/responses.csv")
datos.head(n=5)
```

```python
#Código para crear un dataframe solo con los campos de variables Música y Demografía

variables_musica = list(datos.columns[0:10])
variables_demografía = list(datos.columns[-10:0])
lista_variables = variables_musica + variables_demografía
```

## Correlacion Spearman

```python
data_mus_dem = datos[lista_variables]
f,ax = plt.subplots(figsize=(10, 8))
corr = data_mus_dem.corr(method='spearman')
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
      square=True, annot=True, ax=ax)
```

## Correlacion Kendall

```python
data_mus_dem = datos[lista_variables]
f,ax = plt.subplots(figsize=(10, 8))
corr = data_mus_dem.corr(method='kendall')
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
      square=True, annot=True, ax=ax)
```

## Correlacion Pearson

```python
data_mus_dem = datos[lista_variables]
f,ax = plt.subplots(figsize=(10, 8))
corr = data_mus_dem.corr(method='pearson')
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
      square=True, annot=True, ax=ax)
```
