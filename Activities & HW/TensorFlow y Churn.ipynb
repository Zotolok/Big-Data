## 0. Importar datos

```python
import pandas as pd
from google.colab import drive
from google.colab import files
drive.mount('/content/drive')
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/drive/MyDrive/Proyectos_Personales/Tutorial_TensorFlow/Datos/Churn.csv')
```

```python
X = pd.get_dummies(df.drop(['Churn',"Customer ID"], axis=1))
y = df['Churn'].apply(lambda x: 1 if x=="Yes" else 0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

## 1.1. Instalación
Aparentemente, Colab ya tiene las librerías necesarias. Así que se trabajará de una con TensorFlow

## 1.2. Importar

```python
# Importar dependencias de TensorFlow
from tensorflow.keras.models import Sequential, load_model #
from tensorflow.keras.layers import Dense #
from sklearn.metrics import accuracy_score # Para evaluar el modelo
```

## 2. Construir y compilar el modelo

```python
model = Sequential()
model.add(Dense(units=32, activation = 'relu', input_dim=len(X_train.columns))) # Añadir capas a la red neuronal
model.add(Dense(units=64, activation = 'relu')) #
model.add(Dense(units=1, activation = 'sigmoid'))
```

```python
from sklearn import metrics
# Compilar el modelo: Le decimos a TEnsorflow cómo queremos entrenar el model, con qué métricas lo hará y en cuáles se enfocará.
model.compile(loss='binary_crossentropy', optimizer="sgd", metrics='accuracy')
```

## 3. Establecer, predicir y evaluar (entrenar)

```python
model.fit(X_train, y_train, epochs=200, batch_size=32) # Modelo entrando
# X_train y y_train son el dataframe en entrenamiento
# epochs: Qué tan largo será el entrenamiento
# batch_size: Qué tan largo será el "batch" o el arreglo que queremos pasar a través de Tensorflow antes de hacer una actualización

# Lo que se puede ver son los "loss" y el "accuracy" con los cuales se pueden hacer ciertas predicciones.
# Esto es un modelo de entrenamiento terminado. Idealmente, lo que uno quiere ver cuando se entrena es que el "loss" se reduzca y que, idealmente, el "accuracy" se incremente
# En este caso, empezamos con un accuracy de 0.76 y terminamos con uno de .7979
# Y empezamos con un "loss" de .4977 y terminamos con uno de .4186
```

```python
# Hacer predicciones con TensorFlow
y_hat = model.predict(X_test)
y_hat = [0 if val <0.5 else 1 for val in y_hat]
```

```python
y_hat
```

```python
accuracy_score(y_test, y_hat)
```
